```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv("/root/ashoka-horizons-data-science/Week_3/diabetes.csv")
data.head()
```

```python
fig_mpl_hits, ax_mlp_hits = plt.subplots(figsize=(5,3))
ax_mlp_hits.hist(
    data['Age'].values,
    bins=10,
    color='skyblue',
    edgecolor='black'
)
ax_mlp_hits.set_title("Age distribution of PIMA dataset")
ax_mlp_hits.set_xlabel("Age Bins")
ax_mlp_hits.set_ylabel("Frequency")
plt.show()
```

```python
data=pd.read_csv("/root/ashoka-horizons-data-science/Week_3/diabetes.csv")
# Seaborn
fig_sns_hits, ax_sns_hits = plt. subplots(figsize=(5,3))
# Age Data
sns.histplot(
data=data,
x="Age",
bins=15,
color='pink',
kde=False,
ax=ax_sns_hits
)
ax_sns_hits. set_title("Age distribution of PIMA dataset")
ax_sns_hits. set_xlabel("Age Bins")
ax_sns_hits. set_ylabel ("Frequency")
plt.show()

```

```python
# Seaborn
fig_sns_hits, ax_sns_hits = plt.subplots(figsize=(5,3))
# Age Data
sns.kdeplot (
    data=data,
    x="Age",
    color='teal',
)
ax=ax_sns_hits
ax_sns_hits. set_title("Age distribution (KDE) of PIMA dataset")
ax_sns_hits. set_xlabel("Age Bins")
ax_sns_hits. set_ylabel("Frequency")
plt.show()
```
